```python
def read_matrix(file_name):
    matrix_, suffix_non_matches = [], []
    max_ = 0
    with open(file_name, "r") as file_:
        v, w = file_.readline().split()
#         suffix_non_matches = [int(a) for a in file_.readline().split()]
        for line_ in file_:
            matrix_.append([int(a) if a != '-' else -math.inf for a in line_.split()])

    return np.array(matrix_), np.array(suffix_non_matches), ' ' + v, ' ' + w
```

```python
matrix_a, suff_a, v, w = read_matrix("output_a")
matrix_dij, suff_dij, _, _ = read_matrix("output_dij")
matrix_dp, suff_dp, _, _ = read_matrix("output_dp")
```
